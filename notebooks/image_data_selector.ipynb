```python
# define the directory to search in
directory = r'C:\Users\Adnan\Desktop\Data Science\Final Capstone Project\data\lfw_funneled'

# loop through each file in the directory and its subdirectories
for root, dirs, files in os.walk(directory):
    for file in files:
        # check if the file is an image
        if file.endswith('.jpg') or file.endswith('.jpeg') or file.endswith('.png'):
            # read the image using OpenCV
            img_path = os.path.join(root, file)
            img = cv2.imread(img_path)
            gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
            
            # get the dimensions of the grayscale image
            height, width = gray.shape
            
            # create a blank canvas to copy the grayscale image onto
            canvas = np.zeros((height, width, 3), dtype=np.uint8)
            
            # copy the grayscale image onto the canvas
            canvas[:,:,0] = gray
            canvas[:,:,1] = gray
            canvas[:,:,2] = gray
            
            # do something with the image...
            result = model.predict(gray)
            title = f"Result: {result[0]}"
            cv2.imshow(title, canvas)
            
            key = cv2.waitKey(0) & 0xFF
            if key == ord('y'):
                # copy the grayscale portion of the image to on camera folder
                dest_folder = r'C:\Users\Adnan\Desktop\Data Science\Final Capstone Project\data\on_camera'
                cv2.imwrite(os.path.join(dest_folder, file), gray)
            elif key == ord('n'):
                # copy the grayscale portion of the image to off camera folder
                dest_folder = r'C:\Users\Adnan\Desktop\Data Science\Final Capstone Project\data\off_camera'
                cv2.imwrite(os.path.join(dest_folder, file), gray)

            cv2.destroyAllWindows()
```
